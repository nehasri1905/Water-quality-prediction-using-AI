```python
def plot(d, variable):
    dataframe_pie = d[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize=(9,9), autopct='%1.2f%%', fontsize = 10)
    ax.set_title(variable + ' \n', fontsize = 10)
    return n.round(dataframe_pie/d.shape[0]*100,2)

plot(d, 'Class')
```
